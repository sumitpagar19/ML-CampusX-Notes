```python
import pandas as pd

url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day33-handling-mixed-variables/titanic.csv"
df = pd.read_csv(url, on_bad_lines='skip')  # Skip bad lines if any
print(df.head())
```

```python
import numpy as np
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passangers travelling with')
```

```python
df['number_numerical'] = pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

```python
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python

df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
df.head()
```

```python

df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])
df.head(20)
```
