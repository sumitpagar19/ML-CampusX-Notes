```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
from IPython.display import Image
```

```python
img_read = cv2.imread("new_ziland_lake_SAVED_3.jpg",cv2.IMREAD_GRAYSCALE)
retval,img_thresh = cv2.threshold(img_read,100,255,cv2.THRESH_BINARY)
plt.figure(figsize=[18,5])
plt.subplot(121);plt.imshow(img_read,cmap='gray');plt.title("original");
plt.subplot(122);plt.imshow(img_thresh,cmap='gray');plt.title("thresholded");
print(img_thresh.shape)
plt.show()
```

```python
im_read = cv2.imread("piano_sheet_music.webp",cv2.IMREAD_GRAYSCALE)
retval ,img_thresh_gbl_1 = cv2.threshold(img_read,58,255,cv2.THRESH_BINARY)
retval ,img_thresh_gbl_2 = cv2.threshold(img_read,230,255,cv2.THRESH_BINARY)
img_thresh_adp = cv2.adaptiveThreshold(img_read,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,7)
plt.figure(figsize=[18,4])
plt.subplot(221);plt.imshow(img_read,cmap='gray');plt.title("original");
plt.subplot(222);plt.imshow(img_thresh_gbl_1,cmap='gray');plt.title("thresholded(global: 50");
plt.subplot(223);plt.imshow(img_thresh_gbl_2,cmap='gray');plt.title("thresholded(global: 130");
plt.subplot(224);plt.imshow(img_thresh_adp,cmap='gray');plt.title("thresholded");
plt.show()
```

```python
im = cv2.imread('coca_cola.jpg')
assert im is not None ,"file could not be read,check with os.path.exits()"
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
ret , thresh=cv2.threshold(imgray,127,255,0)
counters,heirarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
print(counters)
print(heirarchy)
```

```python
# img = cv2.cvtColor(im,cv2.COLOR_GRAY2RGB)
cv2.drawContours(im,counters,-1,(0,255,0),3)
plt.imshow(im)
plt.show()
# cv2.imshow('counters',imgray)
```
